```python
def analyze(corpus):
    import pylangacq
    reader = pylangacq.read_chat(corpus)
    # reader.info()
    # reader.participants()
    # reader.ages()
    # reader.word_frequencies(participants='CHI')
    reader.info()
    file_paths = reader.file_paths()

    names = [file_path.split('/')[-1].replace('.cha', '') for file_path in file_paths]

    #names
    from collections import Counter
    word_counts = reader.word_frequencies()

    sorted_word_counts = word_counts.most_common()

    # Print the sorted counts
    # for word, count in sorted_word_counts:
    #     print(f"{word}: {count}")

    import nltk
    from collections import Counter

    # Download the part-of-speech tagger
    nltk.download('averaged_perceptron_tagger')

    # Assuming reader.word_ngrams(1) returns a Counter with tuple keys
    counter_data = reader.word_ngrams(1)

    # Flatten the Counter keys to get just the words (without the tuple structure)
    words = [word[0] for word in counter_data.keys()]
    print("number of unique words: ", len(words))

    # Tag words and filter for nouns (NN, NNS, NNP, NNPS)
    tagged_words = nltk.pos_tag(words)
    nouns = {word: counter_data[(word,)] for word, pos in tagged_words if pos in ["NN", "NNS", "NNP", "NNPS"]}

    # Sort nouns by frequency in descending order
    sorted_nouns = dict(sorted(nouns.items(), key=lambda item: item[1], reverse=True))

    # Print the sorted nouns with their counts
    #display(sorted_nouns)

    sorted_words = dict(sorted_word_counts)

    return (sorted_words, sorted_nouns)
```

```python
def load_corpora(dirs):
    # Combine the text from all files in the directories
    combined_text = ' '.join(
        open(file, 'r').read()
        for dir in dirs
        for file in glob.glob(os.path.join(dir, '**', '*.cha'), recursive=True)
    )

    # Tokenize the text and filter out non-English words
    tokenized_words = word_tokenize(combined_text.lower())
    english_words = set(words.words())
    filtered_words = [word for word in tokenized_words if word.isalpha() and word in english_words]

    # Count the occurrences of each word
    word_counts = Counter(filtered_words)
    word_df = pd.DataFrame(word_counts.items(), columns=['Token', 'Count'])
    word_df['Type'] = 'Word'

    # Count the occurrences of each trigram
    word_trigrams = list(trigrams(filtered_words))
    excluded_words = {'date', 'location', 'id', 'mot', 'chi', 'gra'} # some metadata (incomplete)
    filtered_trigrams = [trigram for trigram in word_trigrams if not any(word in excluded_words for word in trigram)]
    filtered_trigram_counts = Counter(filtered_trigrams)
    trigram_df = pd.DataFrame(filtered_trigram_counts.items(), columns=['Token', 'Count'])
    trigram_df['Type'] = 'Trigram'

    # Concatenate the word and trigram DataFrames
    combined_df = pd.concat([word_df, trigram_df], ignore_index=True)

    # Sort the combined DataFrame by count in descending order
    combined_df = combined_df.sort_values(by='Count', ascending=False)

    return combined_df
```

```python
# Load all corpora into a single pandas DataFrame

corpora = {}

corpora['Bates'] = load_corpora(
    ['data/Bates/Free20'] + 
    [f'data/Bates/{x}28' for x in ['Free', 'Snack', 'Story']]
)

corpora['Champaign'] = load_corpora(
    [f'data/Champaign/{x}{y}' for x in [21, 24, 27, 30, 33, 36] for y in ['P', 'X']]
)

# corpora['Garvey'] = load_corpora(
#     ['data/Garvey']
# )

corpora['Hall'] = load_corpora(
    [f'data/Hall/{x}{y}' for x in ['Black', 'White'] for y in ['Pro', 'Work']]
)

corpora['HSLDD'] = load_corpora(
    [f'data/HSLLD/HV1/{x}' for x in ['BR', 'ER', 'MT', 'TP']] +
    [f'data/HSLLD/HV2/{x}' for x in ['BR', 'ER', 'MT', 'TP']] +
    [f'data/HSLLD/HV3/{x}' for x in ['BR', 'ER', 'ET', 'MT', 'RE', 'TP']] +
    [f'data/HSLLD/HV5/{x}' for x in ['BR', 'LW', 'MT']] +
    [f'data/HSLLD/HV7/{x}' for x in ['ET', 'LW', 'MD', 'MT']]
)

# Combine all corpora into a single DataFrame
df = pd.concat(corpora.values(), keys=corpora.keys(), names=['Corpus', 'Index'])

df
```

```python
import pandas as pd

# Filter the DataFrame to include only trigrams
trigram_df = df[df['Type'] == 'Trigram']

# Group by the trigram and sum their counts to handle duplicates
trigram_df = trigram_df.groupby('Token').agg({'Count': 'sum'}).reset_index()

# Display the 200 most common trigrams in the combined dataset
print('Most common trigrams overall:')
print(trigram_df.sort_values(by='Count', ascending=False).head(200))
```
